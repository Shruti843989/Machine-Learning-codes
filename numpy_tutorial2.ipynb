```python
arr1= np.zeros((2,3),dtype = "int64")
print(arr1,arr1.shape)

arr3= np.ones((2,3))
print(arr3,arr3.shape)

arr3= np.ones((2,3), dtype= "int64")
print(arr3,arr3.shape)

arr4= np.full((3,4),100)
print(arr4,arr4.shape)
```

```python
##indentity matrix
arr5 = np.eye(3)
print(arr5,arr5.shape)
```

```python
##range(start,end,step)
arr6 = np.arange(1,11,2)#range elenements
print(arr6,type(arr6),arr6.shape)
```

```python
arr6 = np.linspace(1,100,4)#evenly spaced arrays
print(arr6,type(arr6),arr6.shape)
```

```python
##operating on numpoy 
#reshaping 2 *3 => 3*2
arr7 = np.array([[1,2,3],[4,5,6]])
print(arr7,arr7.shape)

reshape = arr7.reshape((3,2))
print(reshape,reshape.shape)

flatteened = arr.flatten() #2D => 1D
print(flatteened,flatteened.shape)
```

```python
##indexing 
crr = np.array([[1,2,3],[4,5,6]])
print(crr[0][2])
print(crr[1][2])

#fancy indexing
crr2 = np.array([1,2,3,4,5,6])
idx = [0,1,3]
print(crr2[idx])
```

```python
#boolean indexing
crr3 = np.array([1,2,3,4,5,6])
print(crr3[crr3>2])
print(crr3[crr3%2==0])
print(crr3[crr3%2!=0])
```

```python
#slicing
brr2 = np.array([1,2,3,4,5])
print(brr2[1:4])
print(brr2[:4])
print(brr2[::])
print(brr2[::2])
print(brr2[1:])
```

```python
#copy as view
nums = [1,2,3,4,5]
sub_list = nums[1:3]
print(sub_list)
sub_list[0] = 200

print(sub_list)
print(nums)

#no copy in numpy array for slicing 
nums = np.array([1,2,3,4,5])
sub_list = nums[1:3]
print(sub_list)
sub_list[0] = 200

print(sub_list)
print(f"orginal numpyarray without copy {nums}")

#if want to create a copy in numoy array 
nums = np.array([1,2,3,4,5])
sub_list = nums[1:3].copy()
print(sub_list)
sub_list[0] = 200

print(sub_list)
print(f"orginal numpyarray with copy {nums}")
```
