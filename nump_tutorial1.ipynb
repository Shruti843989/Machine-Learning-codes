```python
#time comparison
size = 100_000_00
#python list 
py_list = list(range(size)) 
start = time.time()#give current time value
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")

#numpy array  
np_arr = np.array(py_list)
start2 = time.time()
#vectorization method 
sq_arr = np_arr**2
end2 = time.time()
print(f"python list time = {end2-start2} seconds")
```
