```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r") as f:
          data = json.load(f) 
    return data
```

```python
data = load_data("store_data.json")
print(data)
```

```python
def clean_data(data):
    test_to_num = {"one": 1, "two" : 2,"three" : 3,"four" : 4 , "five" : 5}
    clean_data=[]#clean the data ko store karao
    unique_user = set()#track the unique values clean data main sian
    for user in data:
        #print(user)
            #clean rating
        raw_rating = str(user["rating"]).strip().lower()#strip the spaces yha str isliye 
        ##lga qki jb yai ek br for loop main gai toh int main convert ho gai thi isliye usse dubre lower case main laye
        if(raw_rating in test_to_num):#raw rating agr string hain 
            raw_rating = test_to_num[raw_rating]#number main convert ho gai raw rating sari
      ##  print(raw_rating) is simple print hua hian override nhi toh oveeride kro 
        user["rating"] = raw_rating
        #print(user)

        #handle the missing value 
        raw_age = user.get("age")#if value toh toh value return krega othersie none retunr krega
        if(raw_age == None):
            user["age"]=None
        
        #dedupplicate data
        if(user["name"].strip() in unique_user):
            continue
        unique_user.add(user["name"])
        clean_data.append(user)
    return clean_data

    
```

```python
data=clean_data(data)
##print(data,len(data))
```

```python
def avg_rating(data):
    # avg rating 
    tot_rating = 0
    for user in data :
        tot_rating += float(user["rating"])
    avg_rating = tot_rating/ len(data)
    print(avg_rating)

    ##percentage of useer with poor rating 
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating +=1
    print(f"% of user with poor rating = { poor_rating/len(data) *100}%")
    
```

```python
avg_rating(data)
```

```python
#build recommendation features
def get_recom(data):
    rec = []
    for user in data:
        curr_rec = {}
        curr_rec["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_rec["brand"] = "Apple"
        else:
            curr_rec["brand"] = "samsung"
        rec.append(curr_rec)
    return rec
```

```python
get_recom(data)
```
