# comprehensive data visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ['A','B','C']
values = [5,7,3]

plt.bar(categories, values)
plt.show()
```

```python
#histrogram
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.show()
```

```python
#Scatter plot\
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
#pie chart
sizes = [40, 30, 20, 10]
labels = ['A','B','C','D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#HeatMAp
# Used for: correlation matrix
data = np.random.rand(5,5)
sns.heatmap(data, annot=True)
plt.show()
```

```python
#BOX PLOT
data = [10,20,30,40,100]

sns.boxplot(data=data)
plt.show()
```

```python
# violin plot
sns.violinplot(data=data)
plt.show()
```

```python
#pair plot
df = sns.load_dataset("iris")
sns.pairplot(df)
```

```python
#interactive plot
import plotly.express as px

df = px.data.iris()

fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```

```python
# subplot
fig, axs = plt.subplots(2,2)

axs[0,0].plot([1,2,3],[4,5,6])
axs[0,1].bar([1,2,3],[3,2,1])

plt.show()
```

```python
# Customization 
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title("Custom Plot")
plt.grid(True)
```

```python
#real dataset 
df = sns.load_dataset("tips")

sns.barplot(x="day", y="total_bill", data=df)
plt.show()
```
